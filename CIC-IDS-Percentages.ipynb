```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score

# Load the first dataset for training
train_data = pd.read_csv('/Users/hemanthsai/Downloads/NIDS Dataset/Dataset/Ratio Bias/2018/cic2018_training_30_10.csv')  # Replace with the path to your training dataset
train_data = train_data.drop('1/3/2018 10:08', axis=1)
train_data = train_data.dropna()
train_data = train_data[np.isfinite(train_data).all(1)]
#attack= 50.28%

# Load the second dataset for testing
test_data = pd.read_csv('/Users/hemanthsai/Downloads/NIDS Dataset/Dataset/Ratio Bias/2018/cic2018_testing_30_34.csv')    # Replace with the path to your testing dataset
test_data = test_data.drop('Timestamp', axis=1)
test_data = test_data.dropna()
test_data = test_data[np.isfinite(test_data).all(1)]
#attack = 12.21%






# Assuming 'target' is the column you want to predict
X_train = train_data.drop('Infilteration', axis=1)
y_train = train_data['Infilteration']

X_test = test_data.drop('Label', axis=1)
y_test = test_data['Label']

# Split the training data for model evaluation
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Define a dictionary of classifiers
classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM': SVC(kernel='linear', C=1),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=3)
}

# Iterate over classifiers
for clf_name, classifier in classifiers.items():
    # Train the classifier
    classifier.fit(X_train_split, y_train_split)

    # Make predictions on the validation set
    y_val_pred = classifier.predict(X_val_split)

    # Evaluate the model's performance on the validation set
    accuracy_val = accuracy_score(y_val_split, y_val_pred)
    f1_val = f1_score(y_val_split, y_val_pred, average='weighted')

    print(f'{clf_name} Accuracy on Validation Set: {accuracy_val:.2f}')
    print(f'{clf_name} F1 Score on Validation Set: {f1_val:.2f}')

    # Test the trained model on the separate test dataset
    y_test_pred = classifier.predict(X_test)

    # Evaluate the model's performance on the test set
    accuracy_test = accuracy_score(y_test, y_test_pred)
    f1_test = f1_score(y_test, y_test_pred, average='weighted')

    print(f'{clf_name} Accuracy on Test Set: {accuracy_test:.2f}')
    print(f'{clf_name} F1 Score on Test Set: {f1_test:.2f}')
    print('\n')
```
